```python
# importing the modules
import requests
from bs4 import BeautifulSoup

url = "https://opentender.eu/at/search/tender"

# creating requests object
html = requests.get(url).content

# creating soup object
data = BeautifulSoup(html, 'html.parser')
table1= data.find('table',id="table-top")
table1
# selects= data.find_all("g", {"class": "tick"})
# # print(type(selects),selects)
# for title in data.find_all('title'):
#   print(title.get_text())

```

### Heading of the table available in the webpage inside a table "table-top"

```python
import pandas as pd

headers = []
for i in table1.find_all('th'):
 title = i.text.strip()
 headers.append(title)
mydata = pd.DataFrame(columns = headers)
mydata
```

### Filter of data on the basis of varied field like "bid price", "Tender Link", "Title" and all the avilable heading. 

```python
from tables import index
# Create a for loop to fill mydata
link_data=[]
for j in table1.find_all('tr')[1:]:
 row_data = j.find_all('td')
 for td in row_data:
   if td.a:
     link_data.append("https://opentender.eu"+str(td.a.get('href')))
 row = [i.text.strip() for i in row_data]
 length = len(mydata)
 mydata.loc[length] = row
original_data=list(link_data)
while(len(mydata["Tender Link"])>len(link_data)):
  link_data.append(" ")
# while(len(mydata["Tender Link"])<len(link_data)):
  # mydata["Tender Link"].append(len(mydata["Tender Link"])+1)
mydata["Tender Link"]=list(link_data)
mydata
```
